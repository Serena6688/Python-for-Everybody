```python
import pandas as pd; import numpy as np; import timeit # Import library
df = pd.read_csv('assets/Business_employment_data.csv', index_col=0)  # Read CSV into a dataframe
df.head()  # Checking the first few rows
```

```python
df['period'] = df.index #Copy the indexed data into its own column
df = df.set_index('region_name') # Set indexof the column to region_name
df.head() # Checking the result
```

```python
new_df=df.rename(mapper=str.strip, axis='columns') #Data Cleaning: remove whitespace
new_df.head() #Display the renamed DataFrame
```

```python
df=df[df['period'] == 2022.12] # Only include rows where the 'period' column has a value of 2022.12
df.loc[ [('Northland'), ('Gisborne')]] # List two region to compare their data
```

```python
type(df.loc['Northland']) # Check the data type of the return
```

```python
copy_df = df.loc[ [('Northland'), ('Gisborne')]] # Make a copy of a DataFrame using .copy()
copy_df.drop('Gisborne') #drop the unwanted data
```

```python
df['series_reference'].unique() # Retrieves all the unique values
```

```python
(df.where(df['series_reference']=='BDCQ.SED1RA') # Use Method Chaining to only include rows where the 'series_reference' column is equal to 'BDCQ.SED1RA'
    .dropna() #drops any rows with missing values
    .set_index(['period','filled jobs']) #sets a new index using the 'period'and 'filled jobs' columns
    .rename(columns={'total_earnings': 'Total_Earnings'})) #  renames the 'total_earnings' column to 'Total_Earnings'.
```

```python
def my_approach(): # Defines a function
    global df
    new_df = df[df['series_reference']=='BDCQ.SED1RA']
    new_df.set_index(['period','filled jobs'], inplace=True)
    return new_df.rename(columns={'total_earnings': 'Total_Earnings'})
timeit.timeit(my_approach, number=1) # Msures the execution time
```
